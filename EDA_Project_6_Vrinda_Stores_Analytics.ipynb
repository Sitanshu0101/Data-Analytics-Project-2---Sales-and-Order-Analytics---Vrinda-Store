# **Data Cleaning and organising **

```python
# This indicates that, Our Dataset has no null values
df.isnull().sum()
```

```python
# Dropping the useless Columns
ds = df.drop(['index', 'Cust ID', 'Date', 'SKU', 'Size', 'currency', 'ship-city', 'ship-postal-code', 'ship-country', 'B2B'], axis = 1)
ds.head()
```

```python
# Just checking the outliers, as the data is correct during Excel analysis, thus, we do not have to remove them.

sns.boxplot(data = ds)
```

## **Data Analysis and Visulaization **

```python
# Figuring out the interrelationship (Value - Oriented)
print('The co-relation will be - ', ds.corr())
sns.heatmap(ds.corr(), annot = True)
```
